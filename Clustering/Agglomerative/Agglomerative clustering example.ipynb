```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sys
sys.path.append('../common')
```

```python
from commonfunctions import visualize_cluster_centroids, visualize_dendrogram, visualize_cluster_data
```

```python
# Read the dataset
dataset = pd.read_csv('../Data/data2.csv')
```

```python
X = dataset.iloc[:, [0,1]].values
clusters = [2] * X.shape[0]
visualize_cluster_data(plt, X, clusters, 'Raw data', filename="AggloExampleRawData")
```

## We will demonstrate the agglomerative clustering in the following loop.
At each step, we will merge the nearest clusters based on euclidean distance; till we arrive at final 2 clusters

```python
from sklearn.cluster import AgglomerativeClustering

for c in range(10,1,-1):
    estimator = AgglomerativeClustering(n_clusters = c)
    estimator.fit(X)
    visualize_cluster_centroids(plt, X, estimator.labels_, 'With '+(str)(c)+' clusters', True, filename = 'Agglo' + (str) (10-c).zfill(2))
```
