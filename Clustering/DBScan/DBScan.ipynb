```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import sys
sys.path.append('../common')
from commonfunctions import visualize_cluster_data, visualize_knee_chart
```

```python
# Read the dataset
dataset = pd.read_csv('../Data/cluster1.csv')
```

Visualize the knee chart for the data set

```python
X = pd.read_csv('../Data/cluster1.csv').iloc[:, [1,2]].values

visualize_knee_chart(plt, X, 'Knee Chart', filename='Knee1')
```

From the chart, we can estimate eps = 1.75

```python
from sklearn.cluster import DBSCAN
estimator = DBSCAN(eps = 1.75, min_samples=5)
estimator.fit(X)
total_clusters = len(set(estimator.labels_)) - (1 if -1 in estimator.labels_ else 0) 
visualize_cluster_data(plt, X, estimator.labels_, 'Total clusters = ' + (str)(total_clusters), filename='DBcluster1')
```
